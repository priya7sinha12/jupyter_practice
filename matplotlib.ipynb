```python
xdata = [0, 2, 4, 6, 8, 10, 12, 14]
ydata = [4, 2, 8, 6, 10, 5, 12, 6]
  

plt.plot(xdata, ydata, linestyle='dotted',color='red',linewidth=6)
plt.show()
```

```python
x_num = [0, 2, 4, 6, 8, 10, 12, 14]
y_num = [4, 2, 8, 6, 10, 5, 12, 6]

plt.xlabel("data of x")
plt.ylabel("data of y")
plt.title("graph  representation")
plt.plot(y,marker='o',ms=20,mfc="yellow",mec='orange',linestyle='--',color='brown',linewidth=3)
plt.show()
```

```python
x_num = [0, 2, 4, 6, 8, 10, 12, 14]
y_num = [9, 8, 6, 10, 5, 12, 6,7]

plt.xlabel("data of x",loc="left",fontsize=15,color="blue")
plt.ylabel("data of y",loc="top",fontsize=10,color="green")
plt.title("graph  representation",loc="center",fontsize=20,color="purple")
plt.plot(y,marker='o',ms=20,mfc="yellow",mec='orange',linestyle='--',color='brown',linewidth=3)
plt.show()
```

```python
x_pie=np.array([56,78,45,67,40])
plt.pie(x_pie)
plt.show()
```

```python
x_pie=np.array([56,78,45,67,40])
mycolors=["red","yellow","blue","pink","orange"]
plt.pie(x_pie,colors=mycolors,radius=2)
plt.show()
```

```python
x_pie=np.array([56,78,45,67,40])
mycolors=["beige","grey","cyan","brown","magenta"]
plt.pie(x_pie,colors=mycolors,radius=1.5,shadow=True,frame=True)
plt.show()
```

```python
x_pie=np.array([56,78,45,67,40])
mycolors=["red","yellow","blue","pink","orange"]
mylabel=["cherry","banana","grapes","pear","mango"]
plt.pie(x_pie,colors=mycolors,labels=mylabel)
plt.show()
```

```python
x_pie=np.array([56,78,45,67,40])
mycolors=["red","yellow","blue","pink","orange"]
mylabel=["cherry","banana","grapes","pear","mango"]
plt.pie(x_pie,colors=mycolors)
plt.legend(labels=mylabel)
plt.show()
```

```python
x_pie=np.array([56,78,45,67,40])
mylabel=["cherry","banana","grapes","pear","mango"]
values=[2,1.5,.4,3,1]
plt.pie(x_pie,explode=values)
plt.legend(labels=mylabel)
plt.show()
```
